## 1. Problem Statement

Derive expressions for the changes in internal energy ($\Delta u$), enthalpy ($\Delta h$), and entropy ($\Delta s$) for a gas obeying the modified equation of state
$$P(v - a) = RT$$
under **isothermal** conditions.

**Given:**
- Equation of state: $P(v - a) = RT$
- Process type: isothermal (constant $T$)
- Variables: $a, P_1, P_2, R, T$

**Required:** Derive symbolic expressions for:
1. $\Delta u$  â€” Change in internal energy
2. $\Delta h$  â€” Change in enthalpy
3. $\Delta s$  â€” Change in entropy

## 2. Schematic

This problem does not involve a physical schematic with components or flow devices. Instead, it concerns a **conceptual thermodynamic system** characterized by a modified gas equation of state:

$$P(v - a) = RT$$

The constant $a$ accounts for the excluded volume effect or effective molecular size in the gas. The system is closed, containing a fixed mass of gas undergoing a **reversible, isothermal** process between two pressures $P_1$ and $P_2$.

The volume change can be obtained from the equation of state, and property changes are to be derived using differential relations.

## 3. Assumptions and Approximations

- **Closed system:** Mass remains constant throughout the process.
- **Isothermal process:** $T = \text{constant}$, implying $dT = 0$.
- **Reversible process:** Ensures applicability of thermodynamic differentials.
- **Equation of state:** $P(v - a) = RT$ holds at all states.
- **Negligible kinetic and potential energy changes:** $\Delta ke \approx 0$, $\Delta pe \approx 0$.
- **Thermodynamic relations valid:** Using definitions $h = u + Pv$ and $ds = \frac{\delta Q_{rev}}{T}$.

## 4. Physical Laws and Governing Equations

We begin with the modified equation of state:
$$P(v - a) = RT$$
so that
$$v = \frac{RT}{P} + a.$$

For an **isothermal** process, $T$ is constant. Property differentials can be related as follows.

### (a) Change in Internal Energy, $\Delta u$

From the first law and the definition of internal energy for an isothermal process of this gas:
$$du = c_v dT + \left[T\left(\frac{\partial P}{\partial T}\right)_v - P\right] dv.$$
Since $dT = 0$ and $P(v - a) = RT$, the partial derivative gives
$$\left(\frac{\partial P}{\partial T}\right)_v = \frac{R}{v - a}.$$
Thus,
$$du = \left[T\frac{R}{v - a} - P\right] dv = 0.$$
Hence,
$$\boxed{\Delta u = 0.}$$

### (b) Change in Enthalpy, $\Delta h$

By definition $h = u + Pv$, so at constant temperature and $\Delta u = 0$:
$$\Delta h = v_2 P_2 - v_1 P_1.$$
Using $v = a + RT/P$, we obtain:
$$\Delta h = a(P_2 - P_1).$$

### (c) Change in Entropy, $\Delta s$

For a reversible process,
$$ds = \frac{\delta Q_{rev}}{T} = \frac{du + Pdv}{T}.$$ 
Using $du = 0$ and substituting $v = a + RT/P$, we get:
$$ds = -R\, d(\ln P).$$
Integrating from $P_1$ to $P_2$:
$$\boxed{\Delta s = -R \ln\left(\frac{P_2}{P_1}\right).}$$

````python
## 5. ðŸ’» Properties and Calculations

import sympy as sp

# Define symbols
P1, P2, a, R = sp.symbols('P_1 P_2 a R', real=True, positive=True)

# (a) Î”u = 0 (isothermal, no change in internal energy)
du_expr = sp.Integer(0)

# (b) Î”h = a (P2 - P1)
dh_expr = a * (P2 - P1)

# (c) Î”s = -R * ln(P2 / P1)
ds_expr = -R * sp.log(P2 / P1)

display(du_expr, dh_expr, ds_expr)
print("Î”u =", du_expr)
print("Î”h =", dh_expr)
print("Î”s =", ds_expr)
````

## 6. Summary and Reasoning

For a gas following the modified equation of state $P(v - a) = RT$ under **isothermal** conditions:

$$\boxed{\Delta u = 0}$$
$$\boxed{\Delta h = a (P_2 - P_1)}$$
$$\boxed{\Delta s = -R \ln\left(\frac{P_2}{P_1}\right)}$$

Key insights:
- Internal energy remains constant because $du = 0$ in an isothermal process with this equation of state.
- Enthalpy varies linearly with the pressure difference due to the additive constant $a$ in the effective volume term.
- Entropy decreases logarithmically with increasing pressure, consistent with reversible isothermal compression.

These relationships demonstrate how non-ideal modifications in an equation of state ($v - a$ instead of $v$) influence derived thermodynamic properties while preserving the overall structure of the classical ideal-gas relationships.
