## 3. Assumptions and Approximations
### Diesel Cycle assumptions
* **Air-Standard Assumption**
    * Working fluid is air, which circulates in a closed loop and behaves as an ideal gas.
    * All processes are internally reversible.
    * Combustion processes are replaced by heat-addition processes from an external source.
    * Exhaust processes are replaced by a heat-rejection process, restoring the working fluid to its initial state.
* **Closed system** (no mass in or out).
* **Pure Substance** We only need two independent variables to determine the state.
* **Properties** We can use `pyCalor` to determine properties. 
* **Ideal Gas Model** (Justification: Air at these conditions behaves as an ideal gas, $p v = R T$)
* **Quasi-equilibrium process** This is a necessary assumption if we are going to relate changes in properties to path functions like work, $W$.
* **Neglect changes in kinetic and potential energy**: $\Delta KE\approx0$, $\Delta PE\approx0$. (Justification: The velocities are typically low, or the change in height is negligible compared to the enthalpy change.)
* **Boundary work** is transferred through expansion ($p dV$).
* **Constant Specific Heats** ($c_p$, $c_v=\text{const}$). (Justification: The specific heat property changes are negligible over the process temperature range, simplifying the calculation of $\Delta h$ to $c_p\Delta T$.
* **Enthalpy/Internal Energy Change**: Changes in enthalpy can be determined via $dh = c_p dT$; changes in internal energy can be determined via $du = c_v dT$ (This is a consequence of the ideal gas model.)
* **Work** The only kind of work a simple compressible system may undergo in an internally reversible process is boundary work.
* **Heat transfer** Heat is transferred in and out
* **Internally reversible process** $\left( \frac{\delta Q}{T} \right)_\text{int,rev}=dS$. An internally reversible process is one where all irreversibilities, e.g., friction, temperature differences, etc., occur outside the system's boundaries. Inside the system, the process occurs without any dissipation, maintaining equilibrium and a uniform temperature.
* **Tds Relations** Because the substance is an ideal gas, we can use $Tds$ relations to compute changes in entropy if we assume constant specific heats.

### Comments on the Otto Cycle
The Diesel Cycle consists of four processes: Compression stroke, power stroke, exhaust stroke, and intake stroke. We can simplify these processes by using the **Air Standard Assumption** to be: Isentropic compression, **constant pressure** heat addition, isentropic expansion, and **isochoric** heat rejection. 

## 4. Physical Laws
### Closed Systems
1.  **First Law of Thermodynamics (Energy Balance for a Closed System):**
$$\Delta E_\text{system} = E_\text{in} - E_\text{out} = (W_\text{in} + Q_\text{in}) - (W_\text{out} + Q_\text{out})$$

2.  **Conservation of Mass (Mass Balance for a Closed System):**
    $$\Delta m_\text{CV} = 0$$

3. **The Entropy Changes of Ideal Gases: $Tds$ Relations:**
$$ ds = c_v \frac{dT}{T} + R\frac{dv}{v}$$

$$ ds = c_p \frac{dT}{T} - R\frac{dp}{p}$$

By definition we can write that the changes in the energy of the system only depend on changes in internal energy, kinetic energy, and potential energy. 

$$\Delta E_\text{system} = \Delta U + \Delta KE + \Delta PE $$

Neglect kinetic and potential energy changes:

$$\Delta E_\text{system} = \Delta U $$

$$\Delta U = U_2 - U_1 = (Q_\text{in} - Q_\text{out}) + (W_\text{in} - W_\text{out}) $$

#### Otto Cycle Definitions
The compression ratio, $r$, is defined as the ratio of the piston-cylinder volume at bottom dead center (BDC) divided by the volume at top dead center (TDC):

$$r = \frac{V_1}{V_2}$$

#### Isentropic Compression
Since this is an isentropic process, we can write our $Tds$ relation as:
$$ 0 = c_v \frac{dT}{T} + R\frac{dv}{v}$$

Integrating from state 1 to 2, and assuming constant specific heats:
$$ 0 = c_v \ln \frac{T_2}{T_1} + R \ln \frac{V_2}{V_1}$$

Which from the definition of compression ratio is:
$$ 0 = c_v \ln \frac{T_2}{T_1} + R \ln \frac{1}{r}$$

We can define $T_2$ as:
$$  T_2 = T_1 \left( \frac{1}{r}\right)^{-\frac{R}{c_v}} = T_1 r^{k - 1} $$

### **Isobaric** Heat Addition
**First Law of Thermodynamics (Energy Balance for a Closed System):**
$$\Delta U = (Q_\text{in} - Q_\text{out}) + (W_\text{in} - W_\text{out}) $$

There is work transfer out and heat transfer in, therefore:
$$\Delta U = U_3 - U_2 = Q_\text{in} - W_\text{out} $$

Since it is a constant pressure process, we can write:

$$W_\text{out} = \int_2^3 p dV =  p_2 \int_{V_2}^{V_3}  dV = $$

Therefore, the heat transferred in becomes:

$$Q_\text{in} = U_3 - U_2 + (p_2 \left(V_3 - V_2 \right)) = (U_3 + p_2 V_3) - (U_2 + p_2 V_2) = H_3 - H_2$$

For an ideal gas, changes in enthalpy can be determined by:

$$dH = m c_p dT$$

For a closed system, there is no change in mass, and if we assume constant specific heats, we have:
$$H_3 - H_2 = m c_p (T_3 - T_2)$$

The heat transferred in is then:
$$ Q_\text{in} = m c_p (T_3 - T_2)$$

On a per mass basis:
$$ q_\text{in} = c_p (T_3 - T_2)$$

We also have for an ideal gas with no change in mass:

$$\frac{p_2 V_2}{T_2} = \frac{p_3 V_3}{T_3} $$

For an isobaric process we have:

$$T_3 = T_2 \frac{V_3}{V_2} $$

### Isentropic expansion
Since this is an isentropic process, we can write our $Tds$ relation as:
$$ 0 = c_v \frac{dT}{T} + R\frac{dv}{v}$$

Integrating, and assuming constant specific heats, we can write:
$$  T_4 = T_3 (V_3/V_4)^{k - 1} $$

Recall that $V_4 = V_1$, $V_2 = V_1/r$, and $V_3 = V_2 r_\text{c}$.

### **Isochoric** heat rejection
**First Law of Thermodynamics (Energy Balance for a Closed System):**
$$\Delta U = (Q_\text{in} - Q_\text{out}) + (W_\text{in} - W_\text{out}) $$

There is no work transfer, and heat is only transferred out, therefore:
$$\Delta U = U_4 - U_1 = Q_\text{out}$$

For an ideal gas, changes in internal energy can be determined by:

$$dU = m c_v dT$$

For a closed system, there is no change in mass, and if we assume constant specific heats, we have:
$$U_4 - U_1 = m c_v (T_4 - T_1)$$

Therefore, the heat rejected is:
$$ Q_\text{out} = m c_v (T_4 - T_1)$$

### Net work
The net work is the difference between the work transferred out minus the work transferred in. Since this is a cycle, the net work transferred out is equal to the net heat transferred in. Therefore:

$$W_\text{net} = Q_\text{in} - Q_\text{out} $$

Which we can write as:

$$W_\text{net} = \left[m c_p (T_3 - T_2) \right] - \left[ m c_v (T_4 - T_1)\right] $$

On a per mass basis:

$$w_\text{net} = \left[c_p (T_3 - T_2) \right] - \left[c_v (T_4 - T_1)\right]  $$

### Thermal engine efficiency
The definition of efficiency for a thermal engine is:

$$\eta_\text{th} = \frac{w_\text{net}}{q_\text{in}} = \frac{\left[c_p (T_3 - T_2) \right] - \left[c_v (T_4 - T_1)\right]}{c_p (T_3 - T_2)}$$

$$\eta_\text{th} = \frac{w_\text{net}}{q_\text{in}} = 1 - \frac{c_v (T_4 - T_1)}{c_p (T_3 - T_2)} = 1 - \frac{(T_4 - T_1)}{k(T_3 - T_2)}$$

We can re-write this as:

$$\eta_\text{th} = 1 - \frac{T_1(T_4/T_1 - 1)}{kT_2(T_3/T_2 - 1)}$$

### Cutoff ratio
We now define a new quantity, the cutoff ratio, $r_\text{c}$, which is the ratio of the cylinder volumes after and before the combustion process:

$$r_\text{c} = \frac{V_3}{V_2} $$

With this we can define our thermal efficiency as:

$$\eta = 1 - \frac{1}{r^{k-1}} \left[ \frac{r_c^k - 1}{k \left(r_c - 1 \right)} \right] $$

````python
# 5. Properties (Code Cell)
!pip install pyCalor numpy matplotlib
from pyCalor import thermo as th
# We will determine properties in two ways. One assumes ideal gas. 
# The other approach we will use pyCalor to determine all properties

#substance
substance = 'air'
# compression ratio
r = 18
# cutoff ratio
rc = 2

# state 1
p1_kPa = 101.325
T1_C = 26.7
V1_m3 = 1.917/1000
st1 = th.state(substance,p=(p1_kPa,'kPa'),T=(T1_C,'C'),name='1')
cv_kJperKgPerK = st1.cv
cp_kJperKgPerK = st1.cp
k = cp_kJperKgPerK/cv_kJperKgPerK
R_kJperKgPerK = st1.R
T1_K = T1_C + 273.15
# state 2
T2_K = T1_K*(r**(k-1))
V2_m3 = V1_m3/r
# state 3
T3_K = T2_K * rc
V3_m3 = V2_m3 * rc

# state 4
V4_m3 = V1_m3
T4_K = T3_K*((V3_m3/V4_m3)**(k-1))
````

````python
# 6 Calculations (Code Cell)
qIn_kJperKg = cp_kJperKgPerK*(T3_K - T2_K)
qOut_kJperKg = cv_kJperKgPerK*(T4_K - T1_K)

wnet_kJperKg = qIn_kJperKg - qOut_kJperKg

eta = wnet_kJperKg/qIn_kJperKg

eta_th_Diesel = 1 - (1/(r**(k-1)))*(((rc**k) - 1)/(k*(rc - 1)))

m_kg = (p1_kPa*V1_m3)/(R_kJperKgPerK*T1_K)

## Alternative solution
### We could also solve this using pyCalor
st1 = th.state(substance,p=(p1_kPa,'kPa'),T=(T1_C,'C'),name='1')
st2 = th.state(substance,s=st1.s,v=st1.v/r,name='2')
st3 = th.state(substance,p=st2.p,v=rc*st2.v,name='3')
st4 = th.state(substance,s=st3.s,v=st1.v,name='4')

m_kg_alt = V1_m3/st1.v
qOut_alt_kJperKg = (st4.u - st1.u)
qIn_alt_kJperKg = (st3.h - st2.h)
wNet_alt_kJperKg = qIn_alt_kJperKg - qOut_alt_kJperKg
eta_alt = wNet_alt_kJperKg/qIn_alt_kJperKg

print(f"(a) T1_K= {T1_K:.2f} K")
print(f"(a) T2_K= {T2_K:.2f} K")
print(f"(a) T3_K= {T3_K:.2f} K")
print(f"(a) T4_K= {T4_K:.2f} K")

print(f"(b) Net work: {wnet_kJperKg*m_kg:.2f} kJ")
print(f"(b) Efficiency (Ideal Gas): {eta*100:.2f} %")

print(f"(c) Alternative solution: Net work: {wNet_alt_kJperKg*m_kg_alt:.2f} kJ")
print(f"(c) Alternative solution: Efficiency: {eta_alt*100:.2f} %")

print(f"(d) eta_th_Diesel: {eta_th_Diesel*100:.2f} %")

#pv and Ts diagrams
pr = th.process([(st1,st2),(st2,st3),(st3,st4),(st4,st1)])
pr.plot("pv");
pr.plot("Ts");

error = (eta - eta_alt)/eta_alt

print(f"eta Error: {error*100:.2f} %")
````
