## 1. Problem Statement

A gas turbine for an automobile is designed with a regenerator. Air enters the compressor of this engine at pressure $p_1$ and temperature $T_1$. The compressor pressure ratio is $r_p = p_2 / p_1$, and the maximum cycle temperature is $T_{\max}$. The cold airstream leaves the regenerator $\Delta T$ degrees cooler than the hot airstream at the inlet of the regenerator.

Assume both the compressor and the turbine are **isentropic**.

The cycle produces a net power output of $\dot{W}_{\text{net}}$.

**Given / Parameters**

- Working fluid: air (modeled as an ideal gas with constant specific heats at room temperature)
- Compressor inlet pressure: $p_1$ (kPa)
- Compressor inlet temperature: $T_1$ ($^\circ$C)
- Compressor pressure ratio: $r_p$
- Maximum cycle temperature (turbine inlet): $T_{\max}$ ($^\circ$C)
- Regenerator temperature approach: cold airstream leaving is $\Delta T$ K cooler than hot airstream inlet
- Net power output: $\dot{W}_{\text{net}}$ (kW)
- Compressor and turbine: isentropic
- Air-standard assumptions with constant specific heats ($c_p$, $c_v$, $R$, $k$)

**Required**

1. Determine the **rate of heat addition** to the cycle, $\dot{Q}_{\text{in}}$ (kW).
2. Determine the **rate of heat rejection** from the cycle, $\dot{Q}_{\text{out}}$ (kW).

## 2. Schematic

We consider an **air-standard Brayton gas turbine cycle with regeneration** for an automobile application.

- **State 1**: Air enters the compressor at low pressure $p_1$ and temperature $T_1$.
- **Compressor (1â†’2)**: Air is compressed isentropically to $p_2 = r_p p_1$ and $T_2$.
- **Regenerator (cold side, 2â†’5)**: Compressed air at state 2 flows through the cold side, preheated to state 5 using turbine exhaust heat.
- **Combustor / Heater (5â†’3)**: Preheated air is further heated at constant pressure to $T_3 = T_{\max}$.
- **Turbine (3â†’4)**: Hot air expands isentropically through the turbine to state 4, producing shaft work.
- **Regenerator (hot side, 4â†’6)**: Turbine exhaust at state 4 passes through the hot side and transfers heat, exiting at state 6.
- **Cooler (6â†’1)**: Exhaust rejects heat and returns to state 1, completing the cycle.

Each component is a steady-flow control volume. Heat transfer occurs in the combustor, regenerator, and cooler; shaft work occurs in the compressor and turbine.

## 3. Assumptions and Approximations

- **Ideal gas model for air:** $p v = R T$ with $R = 0.287$ kJ/kgÂ·K, $c_p = 1.005$, $c_v = 0.718$, $k = 1.4$.
- **Steady flow:** $\dot{m}_1 = \dot{m}_2 = \cdots = \dot{m}_6$.
- **Negligible kinetic and potential energy changes:** $\Delta ke \approx 0$, $\Delta pe \approx 0$.
- **Isentropic compressor and turbine:**
  - $T_2 = T_1 r_p^{(k-1)/k}$
  - $T_4 = T_3 r_p^{(1-k)/k}$
- **Regenerator model:** $T_5 \approx T_4 - \Delta T$ (cold exit is $\Delta T$ below hot inlet).
- **Constant-pressure heat exchangers:** Combustor and cooler.
- **No pressure losses** in regenerator and heat exchangers.

## 4. Physical Laws and Governing Equations

### Compressor (1â†’2)
$$\dot{W}_{\text{comp}} = \dot{m} c_p (T_2 - T_1)$$

### Turbine (3â†’4)
$$\dot{W}_{\text{turb}} = \dot{m} c_p (T_3 - T_4)$$

### Regenerator
$$c_p (T_4 - T_6) = c_p (T_5 - T_2)$$

### Heat Addition / Rejection
$$q_{\text{in}} = c_p (T_3 - T_5), \qquad q_{\text{out}} = c_p (T_6 - T_1)$$

### Net Work and Efficiency
$$w_{\text{net}} = q_{\text{in}} - q_{\text{out}} = c_p[(T_3 - T_5)-(T_6 - T_1)]$$
$$\eta_{\text{th}} = 1 - \frac{T_6 - T_1}{T_3 - T_5}$$

Finally,
$$\dot{Q}_{\text{in}} = \frac{\dot{W}_{\text{net}}}{\eta_{\text{th}}}, \qquad \dot{Q}_{\text{out}} = \dot{Q}_{\text{in}} - \dot{W}_{\text{net}}.$$

````python
## 5. ðŸ’» Properties and Calculations

import numpy as np

# Air properties (kJ/kgÂ·K)
cp = 1.005
cv = 0.718
R = 0.287
k = cp / cv

# Given data
p1_kPa = 100.0
T1_C = 30.0
rp = 8.0
Tmax_C = 800.0
deltaT_regen = 10.0
Wnet_kW = 115.0

# Convert temperatures to K
T1 = T1_C + 273.15
Tmax = Tmax_C + 273.15

# Isentropic relations
T2 = T1 * (rp ** ((k - 1) / k))
T3 = Tmax
T4 = T3 * (rp ** ((1 - k) / k))

# Regenerator temperature model
T5 = 0.5 * (T4 + deltaT_regen + T2)
T6 = T5 - deltaT_regen

# Efficiency and heat rates
eta_th = 1 - (T6 - T1) / (T3 - T5)
Qin_kW = Wnet_kW / eta_th
Qout_kW = Qin_kW - Wnet_kW

print(f"Air props: cp={cp:.3f}, cv={cv:.3f}, k={k:.3f}")
print(f"eta_th = {eta_th:.4f}")
print(f"Qin = {Qin_kW:.2f} kW, Qout = {Qout_kW:.2f} kW")
````

## 6. Summary and Reasoning

This analysis evaluated an air-standard regenerative Brayton cycle for an automobile gas turbine. Using given inlet conditions, pressure ratio, regenerator approach, and net output, we found:

$$\boxed{\eta_{\text{th}} = 1 - \frac{T_6 - T_1}{T_3 - T_5}}$$
$$\boxed{\dot{Q}_{\text{in}} = \frac{\dot{W}_{\text{net}}}{\eta_{\text{th}}}}, \quad \boxed{\dot{Q}_{\text{out}} = \dot{Q}_{\text{in}} - \dot{W}_{\text{net}}}$$

The regenerator improves efficiency by reducing required external heat addition. Deviations from ideal behavior (non-isentropic devices, pressure losses, variable specific heats) would slightly lower efficiency in real systems.
