## 1. Problem Statement

A turbojet aircraft flies with a velocity of 260 m/s at an altitude where the air is at 34.5 kPa and -40°C. The compressor has a pressure ratio of 10, and the temperature of the gases at the turbine inlet is 1093°C. Air enters the compressor at a rate of 45 kg/s. 

**Goal:** Utilizing the cold-air-standard assumptions, determine: 
- (a) the temperature and pressure of the gases at the turbine exit,
- (b) the velocity of the gases at the nozzle exit, and
- (c) the propulsive efficiency of the cycle.

## 3. Assumptions and Approximations
### A. General (Cycle-Wide) Assumptions

1.  **Steady-State/Steady-Flow:** For each device in the cycle ($\Delta m_\text{CV}= 0$, $\Delta E_\text{CV}= 0$, $\Delta S_\text{CV}= 0$).
2.  **Cold-Air-Standard Assumption**
    * Working fluid is air, which circulates in a closed loop and behaves as an ideal gas.
    * All processes are internally reversible.
    * Combustion processes are replaced by heat-addition processes from an external source.
    * Exhaust processes are replaced by a heat-rejection process, restoring the working fluid to its initial state.
    * Air has constant specific heats whose values are determined at room temperature.
3.  **Overall Energy:** All changes in **Potential Energy ($\Delta PE$)** are negligible between components.
4.  ***Property Model and Analysis Method:*** **(The chosen Specific Heat Implementation)**
    * **Approach: Constant Specific Heat Model (Simpler):**
        * **Fundamental Model:** The working fluid is modeled as an **Ideal Gas**.
        * **Purpose:** Used for quick, hand-calculable estimates.
        * **Specific Heats:** $c_p$ and $c_v$ are assumed to be **constant** (typically evaluated at $25^\circ \text{C}$).
        * **First Law:** Changes in energy are calculated using **$\Delta h = c_p \Delta T$** and **$\Delta u = c_v \Delta T$**.
    * **Alternative Approach: Variable Specific Heat Model (Accurate):**
        * **Purpose:** Used for increased accuracy by using temperature-dependent properties.
        * **Tools:** Properties ($h$, $s$, $T$, and $p$) are retrieved directly from a **thermodynamic property library** (e.g., **`pyCalor`**) based on the **Air model**.
        * **First Law:** Changes in energy are calculated using **$\Delta h = h_{\text{out}} - h_{\text{in}}$** and **$\Delta u = u_{\text{out}} - u_{\text{in}}$**.
5.  **No Net Shaft Work:** In the *ideal* turbojet, the work developed by the turbine is exactly equal to the work consumed by the compressor: $\dot{W}_\text{turbine, out} = \dot{W}_\text{compressor, in}$.
6.  **Stationary** The aircraft is stationary and the air is moving towards the aircraft.


### B. Component-Specific Assumptions

1.  **Single Inlet and Outlet:** ($\dot{m}_\text{in}$ and $\dot{m}_\text{out}$).
2.  **Conservation of Mass:** For all devices operating at steady-flow, the mass entering is equal to the mass exiting: **$\dot{m}_{\text{in}} = \dot{m}_{\text{out}}$**).
3.  **Conservation of Energy (First Law):** The **Steady-Flow Energy Equation** is applied to determine work and heat transfer.
4.  **Entropy Balance (Second Law):** The **Entropy Balance for a Control Volume Equation** is applied to determine changes in entropy.

#### 1-2. Diffuser
* **Adiabatic:** $\dot{Q}=0$.
* **No Shaft Work:** $\dot{W}=0$.
* **Ideal Process:** The ideal process is **Isentropic** ($s_{\text{in}} = s_{\text{out}}$). This is because it's reversible and adiabatic.
* **Velocity at the exit** is negligible ($w_2 \approx 0$).

#### 2-3. Compressor
* **Adiabatic:** $\dot{Q}=0$.
* **Ideal Process:** The ideal process is **Isentropic** ($s_{\text{in}} = s_{\text{out}}$). This is because it's reversible and adiabatic.
* **Actual Process:** The actual work is determined by a specified **Isentropic Efficiency ($\eta_C$)**.
* Changes in kinetic energy are negligible (**$\Delta KE \approx 0$**).

#### 3-4. Combustor / Heat Exchanger
* **Isobaric:** $\Delta p=0$ (constant pressure heat addition).
* **No Shaft Work:** $\dot{W}=0$.
* Changes in kinetic energy are negligible (**$\Delta KE \approx 0$**).

#### 4-5. Turbine
* **Adiabatic:** $\dot{Q}=0$.
* **Ideal Process:** The ideal process is **Isentropic** ($s_{\text{in}} = s_{\text{out}}$). This is because it's reversible and adiabatic.
* **Actual Process:** The actual work is determined by a specified **Isentropic Efficiency ($\eta_T$)**.
* Changes in kinetic energy are negligible (**$\Delta KE \approx 0$**).

#### 5-6. Nozzle
* **Adiabatic:** $\dot{Q}=0$.
* **No Shaft Work:** $\dot{W}=0$.
* **Ideal Process:** The ideal process is **Isentropic** ($s_{\text{in}} = s_{\text{out}}$). This is because it's reversible and adiabatic.
* **Velocity at the inlet** is negligible (${w_5 \approx 0$).

## 4. Physical Laws
### Control Volumes: (General Balance Equations)
1.  **First Law of Thermodynamics (Energy Balance for Control Volume):**
    $$\frac{dE_\text{CV}}{dt} = \dot{E}_\text{in} - \dot{E}_\text{out}$$

2.  **Conservation of Mass (Mass Balance for a Control Volume):**
    $$\frac{dm_\text{CV}}{dt} = \dot{m}_\text{in} - \dot{m}_\text{out}$$
    
3.  **Entropy Balance (Second Law for a Control Volume):**
    $$\frac{dS_\text{CV}}{dt} = \dot{S}_\text{in} - \dot{S}_\text{out} + \dot{S}_\text{gen}$$

### Steady-Flow Analysis
Since each devices operates as a **Steady-Flow** device we can write: 

**Mass Balance:**
$$\dot{m}_\text{in} = \dot{m}_\text{out} = \dot{m}$$

**Energy Balance:**
Applying the **Steady-Flow** condition ($\Delta E_\text{CV} = 0$), we can write:
$$\dot{W}_\text{in} + \dot{Q}_\text{in} + \dot{m} \left(h_\text{in} + \frac{w_\text{in}^2}{2} + g z_\text{in} \right) = \dot{W}_\text{out} + \dot{Q}_\text{out} + \dot{m} \left(h_\text{out} + \frac{w_\text{out}^2}{2} + g z_\text{out} \right)$$

Neglecting potential energy effects:
$$\dot{W}_\text{in} + \dot{Q}_\text{in} + \dot{m} \left(h_\text{in} + \frac{w_\text{in}^2}{2} \right) = \dot{W}_\text{out} + \dot{Q}_\text{out} + \dot{m} \left(h_\text{out} + \frac{w_\text{out}^2}{2} \right)$$

**Entropy Balance:**
$$\dot{S}_\text{gen} = \dot{S}_\text{out} - \dot{S}_\text{in}$$

Assuming that the process is adiabatic and reversible (isentropic), we have:
$$\dot{S}_\text{out} = \dot{S}_\text{in}$$

and

$$\dot{m}_\text{out}s_\text{out} = \dot{m}_\text{in}s_\text{in}$$

Therefore,

$$s_\text{in} = s_\text{out}$$

#### Process 1 to 2: Isentropic flow of an ideal gas in a diffuser

Since there is no heat or work transfer in or out of the diffuser, and knowing that the inlet velocity is much greater than the outlet, we can write our conservation of energy as:
$$h_2 = h_1 + w_1^2/2 $$

Recall for an ideal gas:
$$h_2 - h_1 = c_p (T_2 - T_1)$$

$$\implies c_p (T_2 - T_1) + w_2^2/2 = w_1^2/2 $$

Since the velocity at state 2 is negligible:
$$c_p (T_2 - T_1) = w_1^2/2 $$
$$\implies T_2 = w_1^2/(2c_p) +T_1 $$

For an isentropic process, and using our $TdS$ relations, we can write: 
$$p_2 = p_1 \left(\frac{T_2}{T_1} \right)^{k/(k-1)} $$

#### Process 2 to 3: Isentropic compression of an ideal gas in a compressor

The compression ratio, $r_p$, is defined as:
$$p_3 = r_p p_2 $$

For an isentropic process, and using our $TdS$ relations, we can write: 

$$T_3 = T_2 \left(\frac{p_3}{p_2} \right)^{(k-1)/k} $$

#### Process 3 to 4: Combustion chamber

We are only given information about the temperature at the exit of the combustion chamber, as it enters the turbine, $T_4$.

We assume that there is no pressure drop across the chamber: $p_4 = p_3$ 

#### Process 4 to 5: Isentropic expansion of an ideal gas in a turbine

We assume that there is no shaft work transferred out of the device. Therefore, the work transferred out of the turbine is transferred into the compressor:

$$w_\text{compressor, in} = w_\text{turbine, out}$$

$$\implies h_3 - h_2 = h_4 - h_5$$ 

With this we can determine the pressure and temperature at the turbine exit. Again, assuming air behaves as an ideal gas:

$$c_p (T_3 - T_2) = c_p (T_4 - T_5)$$ 

$$\implies T_3 - T_2 = T_4 - T_5$$ 

$$\implies T_5  = T_4 - T_3 + T_2$$ 

For an isentropic process, and using our $TdS$ relations, we can write: 

$$p_5 = p_4 \left(\frac{T_5}{T_4} \right)^{k/(k-1)} $$

#### Process 5-6 Isentropic expansion of an ideal gas in a nozzle
Energy analysis of the nozzle:

$$h_6 + w_6^2/2 = h_5 + w_5^2/2 $$

We neglect the velocity of the air entering the nozzle:

$$h_6 + w_6^2/2 = h_5$$

$$\implies w_6 = \sqrt{2(h_5 - h_6)}$$

Assuming ideal gas:

$$\implies w_6 = \sqrt{2 c_p (T_5 - T_6)}$$

Using our $TdS$ relations we can write:

$$T_6 = T_5 \left(\frac{p_6}{p_5} \right)^{(k-1)/k} $$

#### Propulsion Efficiency:
The propulsive efficiency of a turbojet engine is the ratio of the propulsive power developed, $\dot{W}_P$, to the total heat transfer rate to the working fluid. We can define the thrust developed by a turbojet engine as the difference in the momentum of hte low-velocity air entering the engine and the high-velocity exhaust gases leaving the engine (Newton's Second Law):

$$F = \left(\dot{m} w \right)_\text{exit} - \left(\dot{m} w \right)_\text{inlet} $$

For an aircraft cruising in still air, the inlet velocity is equal to the aircraft velocity: $w_\text{inlet} = V_\text{aircraft}$. The power developed from the thrust on an engine is called the propulsive power, $\dot{W}_P$, which is the thrust (force) times the distance this force acts on the aircraft per unit time. That is, the thrust times the aircraft velocity: 

$$\dot{W}_P = F  V_\text{aircraft} = \dot{m} (w_\text{exit} - w_\text{inlet}) V_\text{aircraft} $$

The propulsive efficiency is then:
$$\eta_P = \frac{\dot{W}_P}{\dot{Q}_\text{in}} $$

where the heat transfer rate is:
$$ \dot{Q}_\text{in} = \dot{m} (h_4 - h_3) = \dot{m} c_p (T_4 - T_3)$$

````python
# 5. Properties (Code Cell)
!pip install pyCalor
from pyCalor import thermo as th

substance = 'air'
Vaircraft_mPerS = 260
mdot_kgPerS = 45
compressorPressureRatio = 10
pAir_Pa = 34.5e3
Tair_K = -40+273.15
Tturbine_inlet_C = 1093 

# state 1
w1_mPerS = Vaircraft_mPerS
T1_K = Tair_K
p1_Pa = pAir_Pa
st1 = th.state(substance,T=(T1_K,'K'),p=(p1_Pa,'Pa'),name='1')
h1_JperKg = st1.h*1000
# cold air standard assumption
cp_kJperKgperK = st1.cp
cv_kJperKgperK = st1.cv
k = cp_kJperKgperK/cv_kJperKgperK
# convserion
cp_JperKgperK = cp_kJperKgperK*1000

# state 2
T2_K = (w1_mPerS**2)/(2*cp_JperKgperK)+T1_K
p2_Pa = p1_Pa*((T2_K/T1_K)**(k/(k-1)))
h2_JperKg = h1_JperKg+(w1_mPerS**2)/2
st2 = th.state(substance,s=st1.s,h=(h2_JperKg,'J/kg'),name='2')

# state 3
p3_Pa = compressorPressureRatio*p2_Pa
T3_K = T2_K*(p3_Pa/p2_Pa)**((k-1)/k)
st3 = th.state(substance,s=st2.s,p=(p3_Pa,'Pa'),name='3')
h3_JperKg = st3.h*1000

# state 4
p4_Pa = p3_Pa
T4_K = Tturbine_inlet_C+273.15
st4 = th.state(substance,T=(T4_K,'K'),p=(p4_Pa,'Pa'),name='4')
h4_JperKg = st4.h*1000

# state 5
h5_JperKg = h4_JperKg - h3_JperKg + h2_JperKg
st5 = th.state(substance,s=st4.s,h=(h5_JperKg,'J/kg'),name='5')
T5_K = T4_K - T3_K + T2_K
p5_Pa = p4_Pa*(T5_K/T4_K)**(k/(k-1))

# state 6
p6_Pa = pAir_Pa
T6_K = T5_K*(p6_Pa/p5_Pa)**((k-1)/k)
st6 = th.state(substance,s=st5.s,p=(p6_Pa,'Pa'),name='6')
h6_JperKg = st6.h*1000

# cold air standard assumption
#cp_JperKgperK = PropsSI('C','T',273.15+25,'P',101325,fluid); print(cp_JperKgperK)
#k = cp_JperKgperK/PropsSI('O','T',273.15+25,'P',101325,fluid); print(k)
````

````python
# 6 Calculations (Code Cell)
print('-------------- part (a) --------------')
print('(a) the temperature of the gases at the turbine exit (ideal gas): T5 = {0:1.1f} K.'.format(T5_K))
print('(a) the pressure of the gases at the turbine exit (ideal gas): p5 = {0:1.1f} Pa.'.format(p5_Pa))

print('Alternative: (a) the temperature of the gases at the turbine exit (pyCalor): T5 = {0:1.1f} K.'.format(st5.T))
print('Alternative: (a) the pressure of the gases at the turbine exit (pyCalor): p5 = {0:1.1f} Pa.'.format(st5.p*1000))

print('-------------- part (b) --------------')
w6_mPerS = (2*cp_JperKgperK*(T5_K - T6_K))**(0.5)
w6_mPerS_Alt = (2*(h5_JperKg - h6_JperKg))**(0.5)
print('(b) the air velocity at the nozzle exit is (ideal gas): w6 = {0:1.1f} m/s.'.format(w6_mPerS))
print('Alternative (b) the air velocity at the nozzle exit is (pyCalor): w6 = {0:1.1f} m/s.'.format(w6_mPerS_Alt))

print('-------------- part (c) --------------')
qIn_W = mdot_kgPerS*cp_JperKgperK*(T4_K - T3_K)
Wp_W = mdot_kgPerS*(w6_mPerS - w1_mPerS)*Vaircraft_mPerS
eta_P = Wp_W/qIn_W
print('(c) The heat transferred in is (ideal gas): qIn_W = {0:1.1f} W.'.format(qIn_W))
print('(c) The propulsive power is (ideal gas): Wp_W = {0:1.1f} W.'.format(Wp_W))
print('(c) The propulsive efficiency is (ideal gas): eta_P = {0:1.1f} %.'.format(eta_P*100))

qIn_W_alt = mdot_kgPerS*(h4_JperKg - h3_JperKg)
Wp_W_alt = mdot_kgPerS*(w6_mPerS_Alt - w1_mPerS)*Vaircraft_mPerS
eta_P_alt = Wp_W_alt/qIn_W_alt
print('Alternative (c) The heat transferred in is (pyCalor): qIn_W_alt = {0:1.1f} W.'.format(qIn_W_alt))
print('Alternative (c) The propulsive power is (pyCalor): Wp_W = {0:1.1f} W.'.format(Wp_W))
print('Alternative (c) The propulsive efficiency is (pyCalor): eta_P_alt = {0:1.1f} %.'.format(eta_P_alt*100))

error_qIn_W = (qIn_W_alt - qIn_W)/qIn_W_alt
error_eta_P = (eta_P_alt-eta_P)/eta_P_alt
print('Error analysis for qIn_W: error_qIn_W = {0:1.1f} %.'.format(error_qIn_W*100))
print('Error analysis for eta_P: error_eta_P = {0:1.1f} %.'.format(error_eta_P*100))

pr = th.process([(st1,st2),(st2,st3),(st3,st4),(st4,st5),(st5,st6)])
fig1 = pr.plot("Ts");
fig2 = pr.plot("pv");

fig1.savefig("figure_1.pdf")
fig2.savefig("figure_2.pdf")
````

## 7. Summary and Reasoning
We've analyzed the Ideal Jet Propulsion cycle. We saw that there is no net shaft work. All of the shaft work developed by the turbine is transferred into the compressor. The work we analyzed here was the propulsion power and we defined the propulsion efficiency as the ratio of the propulsion power to the heat transferred in. We saw that the error in propulsion efficiency due the cold-air-standard assumption was less than 10% justifying its use for a first analysis.
